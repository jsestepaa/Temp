## Escritura

Al ser un lenguaje dinmico, no es necesario definir anteriormente el tipo de una variable en Python, este es detectado automticamente.

```python
a = 3
b = 3.5
c = True
d = 'Hello World'

print(a, type(a), '\n')
print(b, type(b), '\n')
print(c, type(c), '\n')
print(d, type(d), '\n')
```

Debido a que Pyhon no maneja tcnicamente variables sino apuntadores, es posible redefinir una variable sin problemas

```python
a = 3
a = True
a = 'Honk'
print(a)
```

Como el tipo de la variable es definido automticamente por Python, a veces es necesario realizar un cambio de tipo manual.

<center> tipo_nuevo(variable) </center>

```python
a = 3
print(a, type(a), '\n')

b = float(a)
print(b, type(b), '\n')

c = str(a)
print(c, type(c), '\n')
```
