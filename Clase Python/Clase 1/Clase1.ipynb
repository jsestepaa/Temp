# Tipos en Python

Por defecto, Python reconoce los tipos texto (str), entero (int), flotante (float) y booleano (bool)

### String

* str: Cadena de caracteres que depende de la codificación que se esté manejando.
    * 'Hello World'

### Numérico

* int: Números sin decimales.
    * 15
* float: Números con decimales.
    * 15.00

### Booleano

* bool: Valores usados para comparaciones lógicas.
    * False
    * True

## Escritura

Al ser un lenguaje dinámico, no es necesario definir anteriormente el tipo de una variable en Python, este es detectado automáticamente.

```python
a = 3
b = 3.5
c = True
d = 'Hello World'

print(a, type(a), '\n')
print(b, type(b), '\n')
print(c, type(c), '\n')
print(d, type(d), '\n')
```

Debido a que Pyhon no maneja técnicamente variables sino apuntadores, es posible redefinir una variable sin problemas

```python
a = 3
a = True
a = 'Honk'
print(a)
```

Como el tipo de la variable es definido automáticamente por Python, a veces es necesario realizar un cambio de tipo manual.

<center> tipo_nuevo(variable) </center>

```python
a = 3
print(a, type(a), '\n')

b = float(a)
print(b, type(b), '\n')

c = str(a)
print(c, type(c), '\n')
```

## Tipos de datos agrupados

Python maneja por defecto listas, tuplas y diccionarios

### Listas

* list: Contiene distintos elementos en un mismo objeto, se crea usando [   ] o list()

    * lista1 = ['Concepto', 1, 100.99, True]
    * lista2 = ['Esto', 'es', 'una', 'lista']
    * lista3 = list((3, 4, 5, 6, 3))
    * lista4 = list()

### Tuplas

* tuple: Similar a una lista pero inmutable (veremos que significa eso más adelante), se crea usando () o tuple()
    
    * tupla1 = (2,3,4,5)
    * tupla2 = (1,"Septiembre","2018")
    * tupla3 = tuple(lista1)

### Diccionarios

* dict: Agrupación de llaves y valores en un mismo objeto, se crea usando {}

    * {"Name": ["Real"], "Last Name": ['Person'], "email": ["realperson@notascammer.com.ru"], "edad": [35]}

## Operaciones aritméticas

Si bien es cierto, en python podemos requerir del paquete _math_ para realizar operaciones aritméticas complejas, el lenguaje tambien cuenta con distintas formas de cálculo básico.

Estas son:
* +: Suma
* -: Resta
* \*: Multiplicación
* /: División
* //: Division entera
* %: Módulo
* **: Exponente

Adicional a esto, estos operadores son dinámicos (dependen de las variables que estan operando), por ejemplo:

```python
# + suma dos números
a = 3 + 4
print(a)
```

```python
# pero concatena dos caracteres
a = 'Com' + 'plete'
print(a)
```

```python
# y une dos listas
a = [1, 2, 3] + [4, 5, 6]
print(a)
```

```python
# tambien funciona con la multiplicación
a = 'Honk! ' * 4
print(a) 
```

```python
# pero no con todos los operadores
a = 'Explode ' - 'now'
```

* Vectores: Se considera un vector aquel conjunto de elementos que son todos números, cuya dimensión es 1xn o nx1.

* Matrices: Una matriz es una colección de elementos algebraicos a los que se pueden aplicar distintas operaciones númericas, cuya dimensión es mxn.

Ejemplos:
* vector1 = np.array([2,3,4])
* matriz1 = np.array([[1,2,3],[3,4,5]])
