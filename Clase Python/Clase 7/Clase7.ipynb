### Agrupar por

En estos ejemplos agruparemos por la función groupby que es útil a la hora de realizar calculos por grupos para nuestros datos.

Estas manipulaciones son utiles, pues nos permiten interactuar con las bases de datos, para obtener resultados más rápido.

```python
df_by_year = df.groupby('release_year')
type(df_by_year)
```

```python
# Exploremos el Data frame
df_by_year.describe()
```

```python
# Escojamos un registro
list(df_by_year)[10]
```

```python
# Descriptivas básicas por grupos
df_by_year.describe().head()
```

```python
# Calcular una función por grupos
df_med_by_year = df_by_year.mean()
df_med_by_year.head()
```

```python
# Miremos la indexación del grupo
print(df_med_by_year.index)
```

```python
df2 = df.groupby('release_year')
# Miremos la primera entrada de cada subgrupo
df2.first()
```

```python
# Miremos los registros del subgrupo 2017
df2.get_group(2017)
```

```python
# Si queremos agrupar por dos variables en particular
df2 = df.groupby(['release_year', 'rating']).user_rating_score.sum()
pd.DataFrame(df2)
```

### Split columnas 

En muchos casos podemos tener múltiples variable contenidas/concatenadas en una sola columna, para estas situaciones, debemos separar esa columna, manteniendo la estructura de la tabla

```python
ndf = df["group"].str.split(":", n = 1, expand = True) 
ndf
```

```python
# Agregamos ambas columnas
df["rating_description"]= ndf[0] 
df["year"]= ndf[1] 
df.head()
```

```python
# Eliminemos las columnas sobrantes
df.drop(columns =["ratingdescription", 'release_year', 'group'], inplace = True) 
df.head()
```

### Reshape 
Hay ocasiones que la información no se encuentra en la forma en que la necesitamos, por lo cual, es necesario modificar la tabla, para el caso de un dataframe en Python lo podemos realizar de dos maneras:

* stack()
* melt()

```python
df_nba = pd.read_csv('jugadores_nba.csv',sep=";")
df_nba.dropna(inplace=True)
df_nba.drop_duplicates(inplace=True)
df_nba.head()
```

```python
# Método stack - Pivot table
df_stacked = df_nba.stack()
print(type(df_stacked),'\n\n')
df_stacked.head(18)
```

```python
# Retornemos a la tabla
df_unstacked = df_stacked.unstack() 
print(df_unstacked.head(10))
```

```python
# Método melt - Multiples índices
df_melt = df_nba.melt(id_vars =['Name', 'Team'])  
print(df_melt.head(10),"\n\n")
print(df_nba[['Name', 'Team']].head(10)) 
```

```python
# Tambien podemos transponer la tabla
df_t = df_nba.transpose()
df_t
```
